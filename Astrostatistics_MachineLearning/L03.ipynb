```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import time
import random
import scipy
import scipy.stats
```

```python
#generation from a gaussian distribution
sigma=2
N=100000

gauss =scipy.stats.norm(loc=0,scale=2) #gaussian distribution
sample=gauss.rvs(N)

#histogram of gauss distribution 
plt.hist(sample, density=True, bins=30) #density= true to normalize the distribution
x=np.linspace(-3*sigma,3*sigma,100)
plt.plot(x, gauss.pdf(x))
plt.show()
```

```python
#example for a given N
for i in range(0,N):
    if(sample[i]<0):
        sample[i]=-sample[i] #the integral is in half gaussian, for simmetry the negative values set to positive

#function to integrate
def f(x):
    return (x)**3

#calculation for the mean value = from the gaussian sample, substitute in f(x), take the mean 
mean=2*np.sqrt(np.pi*2)/2*np.mean(f(sample))

print(mean, 2*sigma**4)
```

```python
#do for lots of N
N_tot=np.unique(np.logspace(0,6,100,dtype=int)) #better log space, unique = to have only unique values
print(N_tot)
```

```python
s=0.45 #sigma value
expected_value=2*s**4 #expected value from integral

#funcion for integration
def I(sigma, valN):
    gauss =scipy.stats.norm(loc=0,scale=s) #gaussian distribution
    sample=gauss.rvs(valN) #sample in gaussian 

    for i in range(0,valN):
        if(sample[i]<0):
            sample[i]=-sample[i]
            
    return s*np.sqrt(np.pi/2)*np.mean(f(sample))

mean=[]
for i in range(len(N_tot)):
    mean.append(I(s, N_tot[i]))

mean=np.array(mean)

x_ass=np.linspace(N_tot.min(),N_tot.max(),10000)
y_ass=np.ones(10000)*expected_value

plt.plot(N_tot, mean, label='Integration values')
plt.plot(x_ass, y_ass, label='Expected value')
plt.xlabel('log(N)')
plt.ylabel('log(mean)')
plt.legend()
plt.loglog()
```

```python
#histogram of the mean values
plt.hist(mean,bins=15)
plt.show()
```

```python
error=(mean-expected_value)/expected_value #error values of integration

#error function goes down as 1/rad(N)
def err_f(valN):
    return 1/np.sqrt(valN)
    
y_err=[]
for i in range(len(x_ass)):
    y_err.append(err_f(x_ass[i]))
y_err=np.array(y_err)

plt.plot(N_tot,error,'.', label='Error values')
plt.plot(x_ass, y_err, label='Expected 1/rad(N)')
plt.loglog()
plt.legend()
plt.show()
```

```python
#to check the distribution of integral values at a given N

s=0.45 #sigma value

N_more=10000
mean_more=[]

for i in range((N_more)):
    mean_more.append(I(s, N_more))

mean_more=np.array(mean_more)

plt.hist(mean_more, density=True, bins=100)
plt.show()
```

```python
N_deaths=np.array([0,1,2,3,4]) 
N_group=np.array([109,65,22,3,1])
N_tot=np.sum(N_group)
freq=N_group/N_tot #to have the frequence => number group/ total number

plt.scatter(N_deaths, freq)
plt.ylabel('Frequence')
```

```python
plt.scatter(N_deaths, freq, label='Data')
plt.ylabel('Frequence')

mu=np.linspace(0,4,10)
poisson=[]

for i in range(len(mu)):
    poisson.append(scipy.stats.poisson(mu[i]).pmf(N_deaths))

for i in range(len(poisson)):
    plt.plot(N_deaths, poisson[i], label='$\mu$='+str(np.round(mu[i],2)))

plt.legend()
plt.show()
```

```python
#to find best mu => weght average on frequences
mu_best=np.average(N_deaths, weights=freq)

plt.scatter(N_deaths, freq, label='Data')
plt.plot(N_deaths, scipy.stats.poisson(mu_best).pmf(N_deaths), c='orange', label='$\mu_{best}$='+str(round(mu_best,2)))
plt.legend()
plt.ylabel('Frequence')
```
