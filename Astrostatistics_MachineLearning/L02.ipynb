```python
import numpy as np
import matplotlib.pyplot as plt
import random

N=3 #number of doors used
doors=np.zeros(N)

M=1000 #number of tries
win_f=0
win_s=0
win_n=0

for k in range(0,M):
    car = random.randint(0,N-1) #2 goats e 1 car
    doors[car]=1 #the car's door set to 1  

    #generate the choice of the competitor
    p_fixed=random.randint(0,N-1)

    index=[]

    #save the indexes not with the car
    for i in range(0,N):
        if (i!=car):
            index.append(i)

    #presenter choice different from the car
    if (p_fixed==car): #if the fixed choose the car, preseenter open one of the other
        presenter=random.choice(index)

    if (p_fixed!=car): #if the fixed not choose the car, only other door open
        for i in range(len(index)):
            if (index[i]!=p_fixed):
                presenter=index[i]

    #choice of the switch 
    for i in range(len(doors)):
        if (i!=p_fixed)and(i!=presenter):
            p_switch=i

    #save the indexes different from the open door
    index_new=[]
    for i in range(len(doors)):
        if i!=presenter:
            index_new.append(i)

    #choice of the new competitor
    p_new=random.choice(index_new)

    #save of the wins
    if (p_fixed==car):
        win_f+=1
    
    if (p_switch==car):
        win_s+=1
    
    if (p_new==car):
        win_n+=1

#probabilities values of the different wins
prob=np.ones(3)*[win_f,win_s,win_n]/M
print(prob)
```

```python
#plot for the probabilities
plt.bar([0,1,2],[prob[0],prob[1],prob[2]],tick_label=['fixed', 'switch', 'new'])
plt.title('Win probability')
for y in [1/3,1/2,2/3]:
    plt.axhline(y, ls='dotted',c='black')
plt.show()
```
